```python
from pandas import *
import matplotlib.pyplot as plt
import pandas as pd 
import sys 
import matplotlib
```

```python
year = [2008,2009,2010,2011,2012,2013,2014,2015]
```

```python
df = pd.read_excel('EASY_IPL.xlsx')
df
```

```python
total_tie = []
ball =[]
wicket = []
run_loss = []

for i in year:
    ye = df[df['Year'] == i]
    ye
    
    tie = ye[ye['Winning_Team']=='Match Tied']
    total_tie.append(len(tie))
    
    ballleft = ye[ye['Balls_Remaining'] <= 3]
    ball.append(len(ballleft))
    
    win = ye[ye['Win_Type'] == 'wicket']
    
    wek = win[win['Winning_Margin'] <= 1 ]
    wicket.append(len(wek))
    
    
    run = ye[ye['Win_Type'] == 'run']
    
    runwin = run[run['Winning_Margin'] <= 5]
    run_loss.append(len(runwin))
```

```python
final = []

for i in range(len(year)):
    final.append(total_tie[i] + ball[i] + wicket[i] + run_loss[i])
    
```

```python
final
```

```python
finalset = list(zip(year,final))
finalset
```

```python
df = pd.DataFrame(data = finalset , columns = ['year' , 'No. of close matches'])
df
```
