```python
import matplotlib.pyplot as plt
import pandas as pd 
import matplotlib
import csv

no=int(input("Enter number of columns: "))
colls=[]
i=0
while i<no:
    col=input("Enter the name of column: ")
    colls.append(col)
    i+=1
print(colls)

with open('record.csv', 'a+') as file:
    writer = csv.writer(file)
    writer.writerow([colls[0],colls[1],colls[2],colls[3],colls[4]])


marks=[]

def enterdata():
    d=[]
    name=input("Enter name: ")
    d.append(name)
    m1=int(input("Enter marks of sub1: "))
    d.append(m1)
    m2=int(input("Enter marks of sub2: "))
    d.append(m2)
    m3=int(input("Enter marks of sub3: "))
    d.append(m3)
    m4=int(input("Enter marks of sub4: "))
    d.append(m4)
    print(d)
    marks.append(d)
    c=input("Do you want to enter more data? Y/N: ")
    if c=="y" or c=="Y":
        enterdata()
        
enterdata()
print(marks)

for m in marks:
    with open('record.csv', 'a+') as file:
        writer = csv.writer(file)
        writer.writerow([m[0],m[1],m[2],m[3],m[4]])
name_list = []
avg_marks_list = []


with open('record.csv', 'r+',) as file:
    reader = csv.reader(file)
    for row1 in reader:
        headers = row1
        break
    reader = csv.reader(file)
    next(reader, None)
    for row in reader:
        if len(row) ==0:
            print("empty line")
        else:
            new_list = list(row)
            print (new_list)
            name_list.append(new_list[0])
            avg = (int(new_list[1])+int(new_list[2])+int(new_list[3])+int(new_list[4]))/4
            avg_marks_list.append(avg)
            
student_performance = list(zip(name_list,avg_marks_list))
print(student_performance)

#Creating a dataframe using pandas
df = pd.DataFrame(data = student_performance, columns=['Name', 'Average Marks'])
print(df)

# Creating a graph Matplotlib
x = df['Name']
y = df['Average Marks']
plt.bar(x,y)
plt.show()
```
