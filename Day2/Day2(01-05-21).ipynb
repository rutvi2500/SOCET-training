### List Functions

```python
a, b = 10, 20
print(a, b)

list1, list2 = [123, 'Softvan', 'Innoventa'], [456, 'Nimesh']
print("First list length : ", len(list1))
print("Second list length : ", len(list2))

aTuple = (123, 'Softvan', 'Innoventa', 'Harshal')
aList = list(aTuple)
print(aTuple)
print("List elements : ", aList)

list1 = [11, 56, 78, 32, 45]
print(max(list1))
# 78
print(min(list1))
# 11

aList.append(2009)
print("Updated List : ", aList)

aList = [123, 'Softvan', 'Nimesh', 'Innoventa', 123, 'Nimesh', 'Nimesh']
print("Count for Nimesh : ", aList.count('Nimesh'))

alist = [123, 'Softvan', 'Harshal', 'Innoventa', 123]
bList = [2009, 'Nimesh']
alist.extend(bList)
print("Extended List : ", alist)

alist = [123, 'Softvan', 'Nimesh', 'Innoventa']
print("Index for Nimesh : ", alist.index('Nimesh'))

alist = [123, 'Softvan', 'Nimesh', 'Innoventa']
alist.insert(3, 2009)
print("Final List : ", alist)

alist = [123, 'Softvan', 'Nimesh', 'Innoventa']

print("alist: ", alist.pop())
print(alist)

print("alist : ", alist.pop(2))
print(alist)

alist = [123, 'xyz', 'Nimesh', 'abc', 'xyz']

alist.remove('xyz')
print("List : ", alist)

alist = [123, 'xyz', 'Nimesh', 'abc', 'xyz']
alist.reverse()
print("List : ", alist)

alist = [47.5, 123, 1222.45, 11]
alist.sort(reverse=True)
print("List : ", alist)
```

### Dictionary Functions

```python
dict = {'Name': 'Softvan', 'Employee': 100}
print(len(dict))
# 2

dict = {'Name': 'Softvan', 'Employee': 100}
print(type(dict))
# <class 'dict'>

a = str(dict)
print(a[0])
# {
print(type(str(dict)))
# <class 'str'>

dict = {'Name': 'Softvan', 'Employee': 100}
print(len(dict))
# 2
dict.clear()
print(len(dict))
# 0

dict1 = {'Name': 'Softvan', 'Employee': 100}
abc = dict1.copy()
print(abc)
# {'Name': 'Softvan', 'Employee': 100}

tupl = ('sem1', 'sem2', "sem3")
new_dict = dict.fromkeys(tupl)
print(new_dict)
# {'sem1': None, 'sem2': None, 'sem3': None}

dict = dict.fromkeys(tupl, 8.5)
print(dict)
# {'sem1': 8.5, 'sem2': 8.5, 'sem3': 8.5}

dict = {'Name': 'Softvan', 'Employee': 100}
print(dict.get('Employee'))
# 100

print(dict.get('Education'))
# None

dict = {'Name': 'Softvan', 'Employee': 100, 'Type': 'IT Company'}
print(dict.items())
# dict_items([('Name', 'Softvan'), ('Employee', 100), ('Type', 'IT Company')])

print(dict.values())
# dict_values(['Softvan', 100, 'IT Company'])

print(dict.keys())
# dict_keys(['Name', 'Employee', 'Type'])

dict1 = {'Name': 'Softvan', 'Employee': 100}
dict2 = {'Type': 'IT Company'}
dict1.update(dict2)
print(dict1)
# {'Name': 'Softvan', 'Employee': 100, 'Type': 'IT Company'}
print(dict2)
# {'Type': 'IT Company'}
```
